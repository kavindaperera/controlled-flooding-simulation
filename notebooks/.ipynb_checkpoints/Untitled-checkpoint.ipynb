```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import re
```

```python
df_1 = pd.read_csv('C:\\omnetpp-5.6.2\\samples\\tarp-sim\\results\\dataset_1.csv')
df_2 = pd.read_csv('C:\\omnetpp-5.6.2\\samples\\tarp-sim\\results\\dataset_2.csv')
```

```python
df_1.Module = df_1.Module.str.split('[',expand=True)[1].str.split(']',expand=True)[0]
```

```python
df_2.Module = df_2.Module.str.split('[',expand=True)[1].str.split(']',expand=True)[0]
```

```python
plt.plot(df_1.Module, df_1.nbDataPacketsForwarded,  'bx')
plt.plot(df_2.Module, df_2.nbDataPacketsForwarded,  'rx')
fig = plt.gcf()
fig.set_size_inches(20, 5)
fig.savefig('test2png.png', dpi=300)
```
