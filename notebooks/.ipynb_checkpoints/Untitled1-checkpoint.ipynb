```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import re
import matplotlib.patches as mpatches
```

```python
x_tarp_dd = [0, 50, 100, 150]

y_tarp_dd = [
        (299+194)/(300+299)*100, # 0
        (271+177)/(300+271)*100, # 50
        (299+212)/(300+299)*100, # 100
        (299+212)/(300+299)*100, # 150
    ]

plt.xlabel("Pause Time")
plt.ylabel("PDF%")
plt.yticks(np.arange(0, 102, 2))
plt.grid()

plt.plot(x_tarp_dd, y_tarp_dd,  'b-o') 

# red_patch = mpatches.Patch(color='red', label='The red data')
tarp_dd = mpatches.Patch(color='blue', label='TARP-DD')
nodes = mpatches.Patch(color='white', label='# of Nodes = 70')

plt.legend(handles=[tarp_dd, nodes])

fig = plt.gcf()
fig.set_size_inches(8, 8)
```
